```python
def find_free_port():
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.bind(('', 0))
        return s.getsockname()[1]
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from browsermobproxy import Server
import pandas as pd  # Assuming pandas is used for DataFrame operations
import time

BMP_PATH = '/home/g7/browsermob-proxy-2.1.4/bin/browsermob-proxy'
server = Server(BMP_PATH)

    # Setup BrowserMob Proxy
    

server.start()
proxy = server.create_proxy()
time.sleep(1)  # Wait for network requests to complete


# Setup Chrome with Proxy
chrome_options = Options()
chrome_options.add_argument(f'--proxy-server={proxy.proxy}')
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_experimental_option("prefs", {"profile.default_content_setting_values.images": 2})

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# Create a dictionary to store URLs
 # Mock DataFrame for demonstratio
for index, row in df.iloc[:1000].iterrows():
    name = row["name"]
    url = f'https://forvo.com/word/{name}/'
    try:
        proxy.new_har(name)  # Start a new HAR session named after the entry
        driver.get(url)
        
        WebDriverWait(driver, 3).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "ul.pronunciations-list"))
            )
        
            # Find all play buttons within the article with class 'pronunciations'
        play_buttons = driver.find_elements(By.CSS_SELECTOR, "ul.pronunciations-list .play")
        for play_button in play_buttons:
            
            play_button.click()
            time.sleep(2)  # Wait for network requests to complete
        
        har = proxy.har  # Get HAR data
        results_urls = []
        for entry in har['log']['entries']:
            url2 = entry['request']['url']
            if url2.endswith('.mp3'):  # Corrected condition
                results_urls.append(url2)
        print(f"Added {name}: {results_urls} {index}")

        
        url_dict[name] = results_urls  # Append new URLs

    except Exception as e:
        print(str(e))
    

    

```

```python
normal_dict = dict(url_dict)
```

```python
data = []
for name, urls in normal_dict.items():
    for url in urls:
        data.append({'name': name, 'url': url})
df = pd.DataFrame(data)
df.to_csv('output0.csv', index=False)
```
